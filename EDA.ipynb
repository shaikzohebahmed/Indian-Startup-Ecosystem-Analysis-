- Investors' names have ambiguous records (e.g., 'sequoia' and 'Sequoia Capital India') - `Accuracy`

**Messy Data**

- Convert the Amount values from USD to INR for consistency

```python
import codecs
import re

# Function to clean a string by removing control characters, escape sequences, and multiple spaces
def clean_string(value):
    try:
        decoded_value = value.encode().decode('unicode-escape')  # Convert escape sequences to their corresponding characters
        decoded_value = re.sub(r'[\x00-\x1F]', '', decoded_value)  # Remove control characters
        decoded_value = re.sub(r'\\[xX]..', '', decoded_value)  # Remove remaining escape sequences like \x..
        decoded_value = re.sub(r'\s+', ' ', decoded_value)  # Replace multiple spaces with a single space
        return decoded_value.strip()  # Remove leading and trailing whitespace
    except:
        return value  # Return the original value if decoding fails

# Define the fix_encoding function to clean text data
def fix_encoding(text):
    # Skip URLs
    if re.match(r'^https?://\S+', str(text)):
        return text  # Return the original text if it's a URL
    
    decoded_text = codecs.decode(str(text), 'unicode_escape')  # Convert escape sequences to characters
    cleaned_text = re.sub(r'[^a-zA-Z0-9\s\-\']', '', decoded_text)  # Remove characters that are not letters, digits, spaces, hyphens, or apostrophes
    return cleaned_text  # Return the cleaned text


# Apply the clean_string function element-wise to the entire DataFrame and Create a copy of the cleaned DataFrame
df = df1.applymap(clean_string).copy()

# Iterate through columns and apply fix_encoding only to columns with 'object' (string) dtype
columns = ['Startup Name','Industry Vertical','SubVertical','City  Location']
for col in columns:
    if df[col].dtype == 'object':
        df[col] = df[col].apply(fix_encoding)
```

```python
# Fill NaN values in 'amount' column with '0'
df['amount'].fillna('0', inplace=True)

# Replace multiple patterns using a single pass of regular expressions
replace_patterns = [
    (r'[,+]', ''),           # Remove commas
    (r'\++', ''),            # Remove plus signs
    (r'N/A|unknown', '0'),   # Replace 'N/A' and 'unknown' with '0'
    (r'undisclosed', '0' )  # Replace 'undisclosed' (case-insensitive) with '0'
]

for pattern, replacement in replace_patterns:
    df['amount'] =df['amount'].str.replace(pattern, replacement, flags=re.IGNORECASE)
```

```python
df['amount'] = df['amount'].astype('float')
```

```python
def to_inr(dollar):
    inr = dollar * 82.5
    return inr/10000000
```

```python
df['amount'] = df['amount'].apply(to_inr)
```

```python
#function checks for error in format of date column in funding dataframe
def rem_err_date(x):
    if re.search('.',x):#data column has formatting errors like '12/052015','13/042015' where backslash (/) is missing or at wrong position
        return x.replace('.','')
    return x

# Apply the custom function to fix formatting errors in the date column
df['date'] = df['date'].apply(rem_err_date)

# # Replace specific known errors
error_replacements = {
    '12/052015': '12/05/2015',
    '01/07/015': '01/07/2015',
    '15/012015': '15/01/2015',
    '22/01//2015': '22/01/2015',
    '13/042015': '13/04/2015',
    '05/072018': '05/07/2018'
}

df['date'].replace(error_replacements, inplace=True)

# Convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')
```

```python
investors_list = []

for _, row in df.iterrows():
    investors = row['investors'].split(',')
    investors = [investor.strip() for investor in investors]
    data = {
        'date': row['date'],
        'startup': row['startup'],
        'vertical': row['vertical'],
        'subvertical': row['subvertical'],
        'city': row['city'],
        'amount': row['amount'],
        'year': row['year'],
        'month': row['month']
    }
    
    for investor in investors:
        data['investors'] = investor
        investors_list.append(data.copy())
```

```python
new_df = pd.DataFrame(investors_list)
```

```python
# Top investors
top_investors = new_df.groupby('investors')['amount'].sum().sort_values(ascending=False).reset_index()

# Sum investment amounts for both 'SoftBank Group' and 'Softbank'
softbank_group_amount = top_investors.loc[
    top_investors['investors'].isin(['Softbank Group', 'Softbank']), 'amount'
].sum()

# Update the investment amount for 'SoftBank Group'
top_investors.loc[
    top_investors['investors'] == 'Softbank Group', 'amount'
] = softbank_group_amount

# Drop the rows corresponding to 'Softbank'
top_investors = top_investors[top_investors['investors'] != 'Softbank']
```
